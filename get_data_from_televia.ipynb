```python
all_tags_df = []
range_dates = pd.date_range(start='01/01/2024', end='30/05/2024', freq='M')
for tag in tags:
    for date in range_dates:
        start_of_month = date.to_pydatetime()
        start_of_month = start_of_month.replace(day=1)
        start_of_month = start_of_month.strftime('%d/%m/%Y')
        end_of_month = date.to_pydatetime()
        end_of_month = end_of_month.strftime('%d/%m/%Y')
        df = fetch_data_from_televia(tag, start_of_month, end_of_month)
        all_tags_df.append(df)

df = pd.concat(all_tags_df)
```
